```python
# importing required modules
import pandas as pd
from tqdm import tqdm
import numpy as np
```

```python
# reading data
df = pd.read_csv("twitch_gamers\large_twitch_edges.csv")
```

```python
df
```

Number of vertices

```python
# Extracting unique values from both the columns
arr1 = df['numeric_id_1'].unique()
arr2 = df['numeric_id_2'].unique()
lst =  list(set(arr1)  | set(arr2))
len(lst)
```

Number of edges

```python
len(df)
```

Degree of each vertex

```python
# initializing dict
dict = {}
# setting all the vertices as keys and value as 0
for i in range(len(lst)):
    dict[lst[i]] = 0
# increasing the count of vertex present by 1
for i in range(len(df)):
    dict[(df['numeric_id_1'].loc[i])] += 1
```

```python
dict
```

Adjacency list

```python
# initializing dict
dict_matrix = {}
# setting all the vertices as keys and value as a list
for i in range(len(lst)):
    dict_matrix[lst[i]] = []
# appending the adjacent nodes to list
for i in range(len(df)):
    dict_matrix[(df['numeric_id_1'].loc[i])].append((df['numeric_id_2'].loc[i]))
```

```python
dict_matrix
```

Adjacency matrix

```python
adj_lst = []
for i in tqdm(range(len(lst))):
    if i==200:
        break
    temp_lst = []
    for j in range(len(lst)):
        if lst[j] in dict_matrix[lst[i]]:
            temp_lst.append(1)
        else:
            temp_lst.append(0)
    adj_lst.append(temp_lst)
```

```python
adj_lst
```

```python
# For testing that the adjacency matrix is correct if the count matches the respective degree of that node
count = 0
for i in range(len(adj_lst[1])):
    if adj_lst[1][i]==1:
        count += 1
    else:
        continue
count
```
