```python
# Adjacency list of the graph
graph = {
    "Oradea": ["Sibiu", "Zerind",], "Zerind":["Oradea", "Arad",],
    "Arad":["Zerind", "Sibiu", "Timisoara",], "Sibiu":["Oradea", "Arad", "Fagaras", "Rimnicu Vilcea",],
    "Fagaras":["Sibiu", "Bucharest",], "Timisoara": ["Arad", "Lugoj",],
    "Lugoj": ["Timisoara", "Mehadia",], "Mehadia": ["Lugoj", "Drobeta",], 
    "Drobeta": ["Mehadia", "Craiova",], "Craiova": ["Drobeta", "Rimnicu Vilcea", "Pitesti",],
    "Rimnicu Vilcea": ["Craiova", "Sibiu", "Pitesti",], "Pitesti": ["Rimnicu Vilcea", "Craiova", "Bucharest",],
    "Bucharest": ["Pitesti", "Giurgiu", "Fagaras", "Urziceni",], "Giurgiu": ["Bucharest",],
    "Urziceni": ["Bucharest", "Hirsova", "Vaslui",], "Hirsova": ["Urziceni", "Eforie",],
    "Eforie": ["Hirsova",], "Vaslui": ["Urziceni", "Iasi",],
    "Iasi": ["Vaslui", "Neamt",], "Neamt": ["Iasi"],
}
graph
```

```python
visited_list = []
func_g = 0
path_list = []
track_dict = {}
var = True
list_all_paths = []
path_count = 0
final_values = []
shortest_path = []
result = ''
def A_star(node1, node2):
    """
    Returns shortest path between node1 and node2 by A* Search algorithm
    
    Parameter node1: is the starting node of a graph
    Precondition: node1 is a valid node.

    Parameter node2: is the ending node of a graph
    Precondition: node2 is a valid node.
    """
    global visited_list, func_g, path_list, var, list_all_paths, path_count, track_dict, result
    if var:
        if node1==node2:
            visited_list.append(node1)
            path_list.append(node1)
            list_all_paths.append(path_list)
            final_values.append(func_g)
            path_count += 1
            j = 0
            keys_dict = list(track_dict)
            
            while j<len(keys_dict):
                i = (keys_dict[j], track_dict[keys_dict[j]])
                j += 1
                if i[1][1]<min(final_values):
                    if i[0].split('_')[1] not in visited_list:
                        restart_node = i[0].split('_')[1]
                        already_visited_node = i[0].split('_')[0]
                        path_list = path_list[:path_list.index(already_visited_node)+1]
                        func_g = i[1][0]+df[(df['node1'] == already_visited_node) & (df['node2'] ==restart_node)]["weight"].values[0]
                        A_star(restart_node, node2)
            result = list_all_paths[final_values.index(min(final_values))]
            var = False
        elif node1 not in visited_list:
            visited_list.append(node1)
            path_list.append(node1)
            shortest_fn = 0
            nearest_node = ''
            temp_var = True
            for i in range(len(graph[node1])):
                if graph[node1][i] not in path_list:
                    if temp_var:
                        temp_var = False
                        nearest_node = graph[node1][i]
                        shortest_fn = func_g +df[(df['node1'] == node1) & (df['node2'] == graph[node1][i])]["weight"].values[0]+ Heuristic_dist["H_func"][Heuristic_dist["node1"].index(nearest_node)]
                        track_dict[f"{node1}_{graph[node1][i]}"] = (func_g, shortest_fn)
                        nearest_node = graph[node1][i]
                    else:
                        temp_cal = func_g + df[(df['node1'] == node1) & (df['node2'] == graph[node1][i])]["weight"].values[0] + Heuristic_dist["H_func"][Heuristic_dist["node1"].index(graph[node1][i])]
                        if temp_cal<shortest_fn:
                            shortest_fn = temp_cal
                            nearest_node = graph[node1][i]
                        track_dict[f"{node1}_{graph[node1][i]}"] = (func_g, temp_cal)
                    # print(track_dict)
            func_g += df[(df['node1'] == node1) & (df['node2'] == nearest_node)]["weight"].values[0]
            A_star(nearest_node, node2)
A_star("Arad", "Bucharest")
result
```
